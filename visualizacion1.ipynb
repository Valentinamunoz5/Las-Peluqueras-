```python
import matplotlib.pyplot as plt


anios = ['2019', '2020', '2021', '2022', "2023"] #Años
grupo_A = [49, 51, 48, 48, 46]  # Porcentajes para el grupo A en cada año
grupo_B = [51, 49, 53, 52, 54]  # Porcentajes para el grupo B en cada año



# Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 5) )
 # Tamaño opcional para la figura (ancho, alto)

# Barras para cada grupo de personas
plt.bar(anios, grupo_A, label='Hombres', color='LightSalmon')
plt.bar(anios, grupo_B, bottom=grupo_A, label='Mujeres', color='OrangeRed')

# Personalizar el gráfico
plt.xlabel('Género')   # Etiqueta del eje x
plt.ylabel('Porcentaje')      # Etiqueta del eje y
plt.title('Género de consumidores online en Chile')  # Título del gráfico
plt.ylim(0, 100)  # Ajustar el rango del eje y de 0 a 100%
plt.legend()  # Mostrar leyenda

# Mostrar el gráfico
plt.show()


```
